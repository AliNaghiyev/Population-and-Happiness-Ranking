```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
happy_2015=pd.read_csv('2015.csv')
happy_2016=pd.read_csv('2016.csv')
happy_2017=pd.read_csv('2017.csv')
happy_2018=pd.read_csv('2018.csv')
happy_2019=pd.read_csv('2019.csv')
```

```python
# Test whether pd reads data correctly
happy_2015.head(3)
```

```python
# Start indexing from 1 instead of 0
happy_2015.index += 1
happy_2016.index += 1
happy_2017.index += 1
happy_2018.index += 1
happy_2019.index += 1
```

```python
# Renaming column names for simplification (Starting from 2015)
happy_2015=happy_2015.rename(columns={
'Happiness Rank':'Happiness_Rank',
'Happiness Score':'Happiness_Score',
'Economy (GDP per Capita)': 'GDP',
'Health (Life Expectancy)': 'Health',
'Trust (Government Corruption)': 'Government_corruption',
'Dystopia Residual':'Dystopia'})
```

```python
# Rename column for 2016
happy_2016=happy_2016.rename(columns={
'Happiness Rank':'Happiness_Rank',
'Happiness Score':'Happiness_Score',
'Economy (GDP per Capita)': 'GDP',
'Health (Life Expectancy)': 'Health',
'Trust (Government Corruption)': 'Government_corruption',
'Dystopia Residual':'Dystopia'})
```

```python
# Rounding long decimals to 1 point decimals for 2017
happy_2017['Happiness.Score']=happy_2017['Happiness.Score'].round(1)
happy_2017['Whisker.high']=happy_2017['Whisker.high'].round(1)
happy_2017['Whisker.low']=happy_2017['Whisker.low'].round(1)
happy_2017['Economy..GDP.per.Capita.']=happy_2017['Economy..GDP.per.Capita.'].round(1)
happy_2017['Health..Life.Expectancy.']=happy_2017['Health..Life.Expectancy.'].round(1)
happy_2017['Trust..Government.Corruption.']=happy_2017['Trust..Government.Corruption.'].round(1)
happy_2017['Dystopia.Residual']=happy_2017['Dystopia.Residual'].round(1)
happy_2017['Generosity']=happy_2017['Generosity'].round(1)
happy_2017['Family']=happy_2017['Family'].round(1)
happy_2017['Freedom']=happy_2017['Freedom'].round(1)
```

```python
# Rename columns for 2017
happy_2017=happy_2017.rename(columns={
'Happiness.Rank':'Happiness_Rank',
'Happiness.Score':'Happiness_Score',
'Economy..GDP.per.Capita.':'GDP',
'Health..Life.Expectancy.':'Health',
'Trust..Government.Corruption.':'Government_corruption',
'Dystopia.Residual':'Dystopia'})
```

```python
# Rename columns for 2018
happy_2018=happy_2018.rename(columns={
'Overall rank':'Happiness_Rank',
'Country or region':'Country',
'Score':'Happiness_Score',
'GDP per capita':'GDP',
'Healthy life expectancy':'Health',
'Perceptions of corruption':'Government_corruption',
'Freedom to make life choices':'Freedom'})

```

```python
# Rename columns for 2019
happy_2019=happy_2019.rename(columns={
'Overall rank#':'Happiness_Rank',
'Country or region':'Country',
'Score':'Happiness_Score',
'GDP per capita':'GDP',
'Healthy life expectancy':'Health',
'Perceptions of corruption':'Government_corruption',
'Freedom to make life choices':'Freedom'})
```

```python
# Add "year" column for all the years (it will be important for further analysis)
happy_2015['Year']=2015
happy_2016['Year']=2016
happy_2017['Year']=2017
happy_2018['Year']=2018
happy_2019['Year']=2019
```

```python
df_happiness=pd.concat([happy_2015,happy_2016,happy_2017,happy_2018,happy_2019])
df_happiness
```

```python
# Drop some unnecessary columns
dropped_columns=['Standard Error',
                 'Lower Confidence Interval',
                 'Upper Confidence Interval',
                 'Dystopia',
                 'Whisker.high',
                 'Whisker.low']
df_happiness=df_happiness.drop(dropped_columns,axis=1)
df_happiness
```

```python
# Merging Family and Social Support
df_happiness['Social_support']=df_happiness['Social support'].fillna(df_happiness['Family'])
df_happiness=df_happiness.drop(['Family','Social support'],axis=1)
df_happiness
```

```python
# Merging Happiness Rank and Overall Rank
df_happiness['Happiness_Rank']=df_happiness['Happiness_Rank'].fillna(df_happiness['Overall rank'])
df_happiness=df_happiness.drop(['Overall rank'],axis=1)
df_happiness
```

```python
# Converting Happiness_Rank column integer type
df_happiness['Happiness_Rank']=df_happiness['Happiness_Rank'].astype(int)
df_happiness
```

```python
# Rounding decimals to 2 for simplification
df_happiness['Happiness_Score']=df_happiness['Happiness_Score'].round(2)
df_happiness['GDP']=df_happiness['GDP'].round(2)
df_happiness['Health']=df_happiness['Health'].round(2)
df_happiness['Freedom']=df_happiness['Freedom'].round(2)
df_happiness['Government_corruption']=df_happiness['Government_corruption'].round(2)
df_happiness['Social_support']=df_happiness['Social_support'].round(2)
df_happiness['Generosity']=df_happiness['Generosity'].round(2)
df_happiness
```

```python
# Making Population file ready for further analysis
pop_df=pd.read_csv('world_population.csv')
df_population=pop_df[['Country/Territory',
                      'Continent',
                      '2015 Population']]
df_population.index += 1
df_population=df_population.rename(columns={'Country/Territory':'Country',
                                            '2015 Population':'2015_Population'})
df_population
```

```python
# Renaming some country names for happiness_population file
df_happiness=df_happiness.replace({'Hong Kong S.A.R., China':'Hong Kong','Macedonia':'North Macedonia','Palestinian Territories':'Palestine','Trinidad & Tobago':'Trinidad and Tobago','Taiwan Province of China':'Taiwan'})
```

```python
# Merging Population and Happiness Ranks
happiness_population=pd.merge(df_happiness,df_population,on='Country',how='left')
happiness_population.index+=1
happiness_population=happiness_population.drop(['Region'],axis=1)
happiness_population
```

```python
happiness_population=happiness_population.dropna(subset=['Continent','Government_corruption'])
happiness_population.index+=1
happiness_population
```

```python
# Final test for data cleaning
happiness_population.isnull().sum()
```

```python
# Converting final df version to CSV file
happiness_population.to_csv('happiness_population.csv',index=False)
```
